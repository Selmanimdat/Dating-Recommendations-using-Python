```python
# separate data into male and female for recomandation
male_profiles = df[df['Gender'] == 'Male']
female_profiles = df[df['Gender'] == 'Female']
```

```python
# mach score for datig recomendations
```

```python
def calculate_match_score(profile1, profile2):
    shared_interests_score = len(set(eval(profile1['Interests'])).intersection(set(eval(profile2['Interests']))))
    age_difference_score = max(0, 10 - abs(profile1['Age'] - profile2['Age']))
    swiping_history_score = min(profile1['Swiping History'], profile2['Swiping History']) / 100
    relationship_type_score = int(profile1['Looking For'] == profile2['Looking For'])
    total_score = shared_interests_score + age_difference_score + swiping_history_score + relationship_type_score
    return total_score
```

```python
# Example usage:
profile1 = male_profiles.iloc[5]
profile2 = female_profiles.iloc[8]
match_score = calculate_match_score(profile1, profile2)
print(f"Match score between User {profile1['User ID']} and User {profile2['User ID']}: {match_score}")
```

```python
# define a function for recomandations
```

```python
import pandas as pd

def recommend_profiles(male_profiles, female_profiles):
    # expand dimensions for broadcasting
    male_profiles_expanded = male_profiles.assign(key=1)
    female_profiles_expanded = female_profiles.assign(key=1)
    # calculate match scores using vectorized operations
    match_scores = (
        calculate_match_score(male_profiles_expanded, female_profiles_expanded)
        .values.reshape(len(male_profiles), len(female_profiles))
    )
    # find the index of the best match for each male profile
    best_match_indices = match_scores.argmax(axis=1)
    # create dataframe of recommendations
    recommendations = pd.DataFrame({
        'Male Profile': male_profiles,
        'Female Profile': female_profiles.iloc[best_match_indices],
        'Match Score': match_scores.max(axis=1)
    })
    return recommendations
```

```python
# Display the top recommendations
for idx, (male_profile, female_profile, match_score) in enumerate(recommendations[:10]):
    print(f"Recommendation {idx + 1}:")
    print(f"Male Profile (User {male_profile['User ID']}): Age {male_profile['Age']}, Interests {male_profile['Interests']}")
    print(f"Female Profile (User {female_profile['User ID']}): Age {female_profile['Age']}, Interests {female_profile['Interests']}")
    print(f"Match Score: {match_score}")
    print()
#from https://python.plainenglish.io/
```
